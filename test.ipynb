```python
import pandas as pd
```

```python
import json
df1 = []
df2 = []

with open('Список 1.txt', mode="r", encoding = "utf-8-sig") as data_file:
    for line in data_file.readlines():
        df1.append(json.loads(line))
columns = list(df1[0].keys())
df1 = list(map(lambda d : list(d.values()), df1))
df1 = pd.DataFrame(df1, columns = columns)
df1['action'] = df1['custom_params'].apply(lambda item : item['action'])
df1['tag'] = df1['custom_params'].apply(lambda item : item['tag'] )
df1 = df1[['userid', 'action', 'tag']]

with open('Список 2.txt', mode="r", encoding = "utf-8-sig") as data_file:
    for line in data_file.readlines():
        df2.append(json.loads(line))
columns = list(df2[0].keys())
df2 = list(map(lambda d : list(d.values()), df2))
df2 = pd.DataFrame(df2, columns = columns)
df2['action'] = df2['custom_params'].apply(lambda item : item['action'])
df2['tag'] = df2['custom_params'].apply(lambda item : item['tag'] )
df2 = df2[['userid', 'action', 'tag']]
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

%config InlineBackend.figure_format = 'svg' 

%matplotlib inline
```
