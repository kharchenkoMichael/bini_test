```python
import pandas as pd
```

```python
import json
df = []
with open('Список 1.txt', mode="r", encoding = "utf-8-sig") as data_file:
    for line in data_file.readlines():
        df.append(json.loads(line))
```

```python
columns = list(df[0].keys())
df = list(map(lambda d : list(d.values()), df))
df = pd.DataFrame(df, columns = columns)
```

```python
df['custom_params'].apply(lambda a : list(a.values()))
```
